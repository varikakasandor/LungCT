## CNN, train, validating

```python
def conv3x3(in_channels, out_channels, stride=1):
    return nn.Conv2d(in_channels, out_channels, kernel_size=3, 
                     stride=stride, padding=1, bias=False)

class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1, downsample=None):
        super(ResidualBlock, self).__init__()
        self.conv1 = conv3x3(in_channels, out_channels, stride)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = conv3x3(out_channels, out_channels)
        self.bn2 = nn.BatchNorm2d(out_channels)
        self.downsample = downsample
        
    def forward(self, x):
        residual = x
        out = self.conv1(x)
        out = self.bn1(out)
        out = self.relu(out)
        out = self.conv2(out)
        out = self.bn2(out)
        if self.downsample:
            residual = self.downsample(x)
        out += residual
        out = self.relu(out)
        return out

class ResNet(nn.Module):
    def __init__(self, block, layers, num_classes=10):
        super(ResNet, self).__init__()
        self.in_channels = 16   # 16
        self.conv = conv3x3(1, 16)  #16 
        self.bn = nn.BatchNorm2d(16) #16
        self.relu = nn.ReLU(inplace=True)
        self.layer1 = self.make_layer(block, 8, layers[0])  #8
        self.layer2 = self.make_layer(block, 32, layers[1], 2) #32
        self.layer3 = self.make_layer(block, 64, layers[2], 2) #64
        self.avg_pool = nn.AvgPool2d(8)
        self.fc = nn.Linear(4096, num_classes)  #64
        
    def make_layer(self, block, out_channels, blocks, stride=1):
        downsample = None
        if (stride != 1) or (self.in_channels != out_channels):
            downsample = nn.Sequential(
                conv3x3(self.in_channels, out_channels, stride=stride),
                nn.BatchNorm2d(out_channels))
        layers = []
        layers.append(block(self.in_channels, out_channels, stride, downsample))
        self.in_channels = out_channels
        for i in range(1, blocks):
            layers.append(block(out_channels, out_channels))
        return nn.Sequential(*layers)
    
    def forward(self, x):
        out = self.conv(x)
        out = self.bn(out)
        out1 = self.relu(out)
        out2 = self.layer1(out1) #out
        out3 = self.layer2(out2)
        out4 = self.layer3(out3)
        out = self.avg_pool(out4)
        out = out.view(out.size(0), -1)
        out = self.fc(out)
        return out #, (out1,out2,out3,out4)
```

```python
from torchsummary import summary
model_base = ResNet(ResidualBlock,[6,6,6],num_classes=1).to(device)
summary(model_base,(1,256,256))
```

```python
def base_train_step(model,device,optimizer,criterion,train_loader):
    for images, labels in train_loader:
        images = images.view(-1, 1, 256, 256)
        labels = labels.view(-1, 1)
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)
        loss = criterion(outputs, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    return
```

```python
def base_model_eval(model,device,test_loader):
    total = 0
    correct = 0
    zz = 0
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)
        images = images.view(-1, 1, 256, 256)
        labels = labels.view(-1, 1)
        outputs = model(images)
        total += labels.size(0)
        raw = outputs>=0.5
        for k in range(labels.size(0)):
            zz = labels[k] >= 0.5
            if raw[k] == zz:
                correct += 1
    return correct,total
```

# AUC and ROC Curve

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot
```

```python
model_weights = []
model_weights.append("states/leftlungaffected_nov27_resnet18_custom_sgd_8.ckpt")
model_weights.append("states/rightlungaffected_nov27_resnet18_custom_sgd_3.ckpt")
model_weights.append("states/cavernsleft_nov27_resnet18_custom_sgd_13.ckpt")
model_weights.append("states/cavernsright_nov27_resnet18_custom_sgd_12.ckpt")
model_weights.append("states/pleurisyleft_nov27_resnet18_custom_sgd_7.ckpt")
model_weights.append("states/pleurisyright_nov27_resnet18_custom_sgd_11.ckpt")
```

```python
#itt annyit változtattam az előző eval-hoz képest, hogy csinál egy pred listát amit össze tudunk vetni a labellel (Ture/False-ból el ez is, és a Label-ös is), amiből ROC és AUC-ot tudunk csinálni

def base_model_eval_2(model,device,test_loader):
    total = 0
    correct = 0
    zz = 0
    pred = []
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)
        images = images.view(-1, 1, 256, 256)
        labels = labels.view(-1, 1)
        outputs = model(images)
        total += labels.size(0)
        raw = outputs>=0.5
        for k in range(labels.size(0)):
            zz = labels[k] >= 0.5
            if raw[k] == zz:
                correct += 1
        if outputs >= 0.5:
            pred.append(True)
        elif outputs < 0.5:
            pred.append(False)
    return correct,total, pred
```
